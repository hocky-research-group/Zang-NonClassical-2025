```python
import numpy as np
e1 = np.array([2.09,1.27,0.45,-0.37,-1.19,-2.01,-2.83])
e2 = -22.89
e = e1 + e2
z1 = np.array([-1.61,-2.41,-3.20,-4.0,-4.8,-5.6,-6.39])
z2 = -20.65
z3 = 0.019
z4 = z1 + z2 +z3
z5 = z1 + z3
v = np.array([-20,-25,-30,-35,-40,-45,-50])
```

```python
import matplotlib
import matplotlib.pyplot as plt
import os
workdir = "/scratch/projects/hockygroup/sp6310/ionic_crystal/ionic_crystal/ExperimentalData/New_simulation2/corrected_pot2/charged_wall/\
fromOldruns"

font = {'family' : 'serif',
        'size'   : 18}
matplotlib.rc('font', **font)
fig = plt.figure(figsize=(8,6))
plt.plot(v,e,marker="o",label = "CsCl-like")
plt.plot(v,z4,marker="o",label="Zanganite")
plt.legend()
#plt.xlim(0,1600000000)
plt.xlabel("Surface potential (mV)")
plt.ylabel("Energy per particle (kBT)")
plt.savefig(os.path.join(workdir,"energyperpart_firstlayer_CsCl_Zanganite_sum3.eps"),bbox_inches='tight',format='eps',dpi=2000)
```
